```python
#bll begins
def add(num1,num2):
    return (num1+num2)

def subt(num1,num2):
    return (num1-num2)
    
def mult(num1,num2):
    return (num1*num2)
    
def divi(num1,num2):
    if(num2!=0):
        return num1/num2
    else:
        raise Exception("divide by zero error")
    
def power(num1,num2):
    return (num1**num2)
    
def remain(num1,num2):
    return (num1%num2)
#bll ends
#pl begins
while(True):
    user_msg=""" 
                 1-addition
                 2-subtaction
                 3-multiplication
                 4-divide
                 5-power
                 6-remainder
                 """
    print(user_msg)
    num1=int(input("enter the first number"))
    num2=int(input("enter the second number"))    
    choice=input("enter the choice")
    choice=int(choice)
    if(choice==1):
        result1=add(num1,num2)
        print("the addition of two numbers is",result1)
    elif(choice==2):
        result2=subt(num1,num2)
        print("the subtraction of two numbers is",result2)
    elif(choice==3):
        result3=mult(num1,num2)
        print("the multiplication of two numbers is",result3)
    elif(choice==4):
        try:
            result4=divi(num1,num2)
            print("the division of two numbers is",result4)
        except Exception as ex:
            print(ex)
    elif(choice==5):
        result5=power(num1,num2)
        print("the power of two numbers is",result5)
    elif(choice==6):
        result6=remain(num1,num2)
        print("the remainder of two numbers is",result6)
    else:
        print("you have entered wrong choice")
#pl ends
```
